# Outliers

- In data analysis, outliers are data points that are significantly different from other data points in a dataset. Outliers are observations that fall far away from the bulk of the data points, and they can have a significant impact on statistical analysis, modeling, and machine learning algorithms.

- Outliers can arise due to various reasons such as measurement or recording errors, data entry errors, or natural variation in the data. Outliers can also occur due to unusual events, anomalies, or extreme values in the data.

- It is important to identify and handle outliers appropriately in data analysis. Outliers can distort statistical measures such as mean and standard deviation and can affect the results of data analysis, making it difficult to draw meaningful insights from the data.

- There are various methods to detect and handle outliers, such as visual inspection of the data using box plots or scatter plots, statistical methods such as z-score or IQR (Interquartile Range), or machine learning methods such as clustering or anomaly detection. The method used to handle outliers will depend on the type of data, the research question, and the context of the analysis.

```python
# define continuous variable and plot

continous_features = ["age", "trestbps", "chol", "thalach", "oldpeak"]

def outliers(data_out, drop = False):
    for each_feature in data_out.columns:
        feature_data = data_out[each_feature]
        Q1 = np.percentile(feature_data, 25) # 25th percentile
        Q3 = np.percentile(feature_data, 75) # 75th percentile
        IQR = Q3 - Q1 # Interquartile range
        outlier_step = 1.5 * IQR # outlier step
        outliers = feature_data[~((feature_data >= Q1 - outlier_step) & (feature_data <= Q3 + outlier_step))].index.tolist()
        if not drop:
            print(f"Feature {each_feature} has {len(outliers)} outliers")
        if drop:
            data_temp.drop(outliers, inplace=True, errors="ignore")
            print(f"Outliers from {each_feature} feature has been removed")

outliers(data_temp[continous_features])
```

## Distribution and Relationship

### a) target variable distribution

```python
print(data_temp["target"].value_counts())

# data_temp["target"].value_counts().plot(kind="bar")

fig, ax = plt.subplots(figsize=(5,4))

name = ["Disease", "No Disease"]

ax = data_temp.target.value_counts().plot(kind="bar")

ax.set_title("Heart Disease Classes", fontsize=13, weight="bold")
ax.set_xticklabels(name, rotation=0)

# to calculate the percentage of each class

totals = []

for i in ax.patches:
    totals.append(i.get_height())
total = sum(totals)

for i in ax.patches:
    ax.text(i.get_x()+0.9, i.get_height()-50, \
             str(round((i.get_height()/total)*100, 2)) + "%",
             fontsize=12, color="white", weight="bold")

plt.tight_layout()
```
